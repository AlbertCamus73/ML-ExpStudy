```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 

from sklearn.neighbors import KNeighborsClassifier
```

```python
XTrain = []
for i in range(1,501):
    XTrain.append(plt.imread('./data/0/0_%d.bmp'%(1)))
```

```python
XTrain = []
XTest = []
YTrain = []
YTest = [] #target 后续作为测试数据标签

for i in range (10):
    for j in range(1,501):
        if j < 451: #保存到训练数据中
            XTrain.append(plt.imread('./data/%d/%d_%d.bmp'%(i,i,j)).reshape(-1))
            YTrain.append(i)
        else: #保存到测试数据
            XTest.append(plt.imread('./data/%d/%d_%d.bmp'%(i,i,j)).reshape(-1))
            YTest.append(i)
```

```python
XTrain,YTrain = np.array(XTrain),np.array(YTrain)
XTest,YTest = np.array(XTest),np.array(YTest)
```

```python
n = np.random.randint(0,501,1)
num = XTest[n].reshape(28,28)
plt.imshow(num,cmap='gray')
```

```python
knn = KNeighborsClassifier()
                            #对训练数据使用KNN训练
knn.fit(XTrain,YTrain)
```

```python
n = np.random.randint(0,501,25)

Y = knn.predict(XTest)
```

```python
    #绘制测试结果对照图
n = np.random.randint(0,501,25)
plt.figure(figsize=(12,16))
ImDatas = XTest[n]
ImTarget = YTest[n]
ImPredict = Y[n]
for i in range(25):
        plt.subplot(5,5,(i+1))
        plt.imshow(ImDatas[i].reshape(28,28),cmap = 'gray')
        plt.title('predict:%d'%(ImPredict[i])+'\n true:%d'%(ImTarget[i]))
```
